```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales_train_compltete_1.csv')
kaggle_test = pd.read_csv('../competitive-data-science-predict-future-sales/test.csv')
```

```python
kaggle_test['date__month'] = 34
kaggle_test['date__month_of_year'] = 11
kaggle_test['date__year'] = 2015
kaggle_test.drop(columns='ID', inplace=True)
```

```python
### USABLE ONLY FOR COLUMN WITH ONLY ONE UNIQUE VAL that depends on month
cols = ['avg_monthly_sales','shop_monthly_avg_sales',
        'holidays_in_month','amt_weekends_month',
        'amt_free_days_month', 'season']
for col in cols:
    kaggle_test[col] = df[df['date__month'] == kaggle_test['date__month'].max() - 12][col].max()
```

```python
len(kaggle_test)
```

```python
cols = ['item_category_id', 'item_name', 'item_category_name',
       'shop_name','category_cnt_month', 'category_monthly_avg_sales',
       'city', 'price_segment', 'shop_age_month',
       'age_group_cat', 'age_group_shop', 'category_shop_age_months',
      'shop_category_cnt_month_avg'
       ]
for col in cols:
    kaggle_test = pd.merge(df[df['date__month'] == 22].groupby(['shop_id','item_id'])[col].last().reset_index(), kaggle_test, on=['shop_id', 'item_id'], how="right")
```

```python
len(kaggle_test)
```

```python
drop_cols = ['cat_by_shop_age_days',
 'category_cnt_month_avg',
 'date',
 'date__day',
 'date__day_of_month',
 'date__day_of_week',
 'date__week',
 'date__week_of_year',
 'first_sale_date',
 'is_holiday',
 'is_weekend',
 'item_cnt_day',
 'item_cnt_week',
 'item_cnt_year',
 'shop_category_cnt_month',
 'total_sales_month',
 'shop_category_cnt_year',
 'category_cnt_year'
 ]

kaggle_test = pd.merge(df[df['date__month'] == 33].groupby(['shop_id','item_id'])['item_price'].last().reset_index(), kaggle_test, on=['shop_id', 'item_id'], how="right")
```

```python
df.drop(columns=drop_cols, inplace=True)
```

```python
kaggle_test.drop(columns=['date__month', 'date__month_of_year'], inplace=True)
```

```python
kaggle_test['date__month'] = 34
kaggle_test['date__month_of_year'] = 11

```

```python
df_con = pd.concat([df, kaggle_test], ignore_index=True)
```

```python
lag_columns = ['amt_free_days_month', 'amt_weekends_month', 'avg_monthly_sales',
 'category_cnt_month', 'category_monthly_avg_sales', 'holidays_in_month',
'shop_monthly_avg_sales']
```

```python
lags = [1, 2, 3, 4, 6, 12]


for col in lag_columns:
    for lag in lags:
        new_column_name = f'{col}_lag_{lag}'
        df_con[new_column_name] = df_con.groupby(['shop_id', 'item_id'])[col].shift(lag)

print('Lags added')
```
